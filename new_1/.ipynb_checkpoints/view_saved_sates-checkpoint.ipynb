```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('states.csv')
```

```python
df.drop(['Unnamed: 0'], axis=1).plot()
```

```python
df[['0', '1']].plot()
# df[['0', '1']].loc[1750:2500].plot()
```

```python
df1 = pd.read_csv('actions.csv')
df1['0'].plot()
```

```python
fig,ax = plt.subplots()
df['0'].loc[:].plot(ax=ax, alpha=0.5, label='coords')
df['1'].loc[:].plot(ax=ax, alpha=0.5, label='angle')
df1['0'].loc[:].plot(ax=ax, alpha=0.5, label='control')
plt.legend()
plt.show()
```

# Angles

```python
fig, axes = plt.subplots(1,2, figsize=(10,3))


st = df.shape[0] - 100
fn = df.shape[0] - 1
(df['0'].loc[st:fn] * 10.75).plot(ax=axes[0])
axes[0].set_title('Distance')
axes[0].set_xlabel('Steps')
axes[0].set_ylabel('Dist, cm?')

(df['1']*180/np.pi).loc[st:fn].plot(ax=axes[1], color='g')
# axes[1].plot(df.index, df['1']/180 * np.pi, color='g')
axes[1].set_title('Inclination')
axes[1].set_xlabel('Steps')
axes[1].set_ylabel('Angle, Deg')
# axes[0,1].set_ylim(-190,190)
# axes[0,1].set_yticks(np.arange(-180,181,20))

plt.show()
```
